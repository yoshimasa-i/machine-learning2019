# プラス成長とマイナス成長の境目を示すグラフ作成

```python
%matplotlib inline
import pandas as pd
```

### CSV読み込み

```python
df = pd.read_csv("population.csv", encoding="SHIFT_JIS")
```

### 増減を調べて並び替える

```python
df["増減"] = df["平成28年"] - df["平成12年"]
df = df.sort_values(by=["増減"], ascending=False)
```

### プラスとマイナスの中間を抜き出す

```python
mid = df[5:15]
```

## グラフのスタイルに ggplot を利用する

```python
import matplotlib
matplotlib.style.use('ggplot')
```

### グラフ描画

```python
plt = mid.plot.bar(y="増減", x="都道府県")
plt.axhline(0, color='k') # 0のラインを強調して完了
```
