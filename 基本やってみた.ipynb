### 人口の少ない順に並び替えて表示

並び替え(ソート)を行うには、sort_valuesメソッドを使うだけ。

```python
import pandas as pd
df = pd.read_csv("population.csv", encoding="SHIFT_JIS")
df.sort_values(by=["平成28年"], ascending=True)
```

### グラフを描画

CSVファイルを元にグラフを描画

```python
%matplotlib inline 
# グラフ化
import pandas as pd  
# pandasを pdに入れる
df = pd.read_csv("population.csv", encoding="SHIFT_JIS") 
#pdの結果をdfに入れる
df.plot(y=["平成28年"])

#dfをbarチャートでプロット、Y横軸に人口表示
```

#### 比較してみる

```python
%matplotlib inline 
# グラフ化
import pandas as pd  
# pandasを pdに入れる
df = pd.read_csv("population.csv", encoding="SHIFT_JIS") 
#pdの結果をdfに入れる
df.plot.bar(y=["平成28年","平成12年"])

#dfをbarチャートでプロット、Y横軸に人口表示
```
